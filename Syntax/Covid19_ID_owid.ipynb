```python
import numpy as np
import pandas as pd 
import wget
import os

url_owid = ['https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv']

path_url = r'C:\Users\Acer\Documents\Belajar Data Science\Source Code Project\Covid19-Report-Indonesia\Dataset\Owid'

for url in url_owid:
    filename= path_url + '/' + os.path.basename(url)
    if os.path.exists(filename):
        os.remove(filename) # if exist, remove it directly
    wget.download(url, out=path_url)
```

```python
cov19_data_owid = pd.read_csv(r'C:\Users\Acer\Documents\Belajar Data Science\Source Code Project\Covid19-Report-Indonesia\Dataset\Owid\owid-covid-data.csv')
cov19_data_owid.head()
```

```python
cov19_id_owid = cov19_data_owid[cov19_data_owid['location'] == 'Indonesia']
cov19_id_owid.tail()
```

```python
print(cov19_id_owid.info())
print(cov19_id_owid.isna().sum())
```

```python
# Handle missing value:
## first using ffill method
cov19_id_owid = cov19_id_owid.fillna(method='ffill')
```

```python
## second, replace null values with 0
cov19_id_owid = cov19_id_owid.fillna(0)
```
